```python
import pandas as pd
```

```python
data = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG (1).csv', encoding='latin-1', error_bad_lines=False)
```

```python
df=pd.DataFrame(data) 
df
```

# What is the total sum of Animal Fat produced in 2014 and 2017 respectively

```python
# What is the total sum of Animal Fat produced in 2014 and 2017 respectively
df.groupby('Item').sum()
```

```python
# What is the total sum of Animal Fat produced in 2014 and 2017
Animal_Fat_2014 = df.groupby('Item').sum().iloc[3, 3]
Animal_Fat_2017 = df.groupby('Item').sum().iloc[3, 6] 
print(f'The total sum of Animal Fat produced in 2014 and 2017 is: {Animal_Fat_2014} and {Animal_Fat_2017}')
```
